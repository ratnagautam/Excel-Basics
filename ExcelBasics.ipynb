Ques1. Explain the difference between Relative , Absolute and Mixed Cell Referencing.
You have the following dataset:

Vehicle           Price
Car              500000
Bicycle           10500
Auto-Rickshaw     25000

    - The difference Between Relative, Absolute and Mixed Cell are as follows:
    A. Relative Cell Reference- It adapts the formulas automatically and also saves the timeIt is convenient whenever you need to repeat the same calculation across multiple rows and columns.
    Eg: If we copy the formula=A1+B1 from row 1 to row 2, the formula will become=A2+B2.

    B. Absolute Cell Reference- To get the same value in other cells too, we use this referenc. Gives the exact/correct answer.We cam use an absolute reference to keep a row or column constant. It is designated in a formula by the addition of a dollar sign before either the row or the column.
    Eg: We will use the relative A2 and absolute $A$2 formats in most formulas.

    C. Mixed Cell Reference   - It gives flexibility in the operation of other cells which includes rows and columns. These are use less frequently.
